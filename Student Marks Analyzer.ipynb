```python
import numpy as np

test1 = np.array([78, 85, 62, 90, 74])
test2 = np.array([82, 88, 70, 92, 80])

total_marks = test1 + test2
print("Total Marks:", total_marks)

average = total_marks / 2
print("Average Marks:", average)

first_three = total_marks[:3]
print("First 3 student marks:", first_three)

print("Highest:", total_marks.max())
print("Lowest:", total_marks.min())
```
